```python
count = [0, 0, 0, 0]
for i in range(len(males_df)):
    if males_df.loc[males_df.index[i], 'status'] == "متزوج":
        if males_df.loc[males_df.index[i], 'area_level'] == "مرتفع":
            count[3] = count[3] + 1
        elif males_df.loc[males_df.index[i], 'area_level'] == "متوسط":
            count[2] = count[2] + 1
        elif males_df.loc[males_df.index[i], 'area_level'] == "شعبى":
            count[1] = count[1] + 1
        else:
            count[0] = count[0] + 1

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
levels = ['Suburbs', 'Rural', 'Medium class', 'High class']
ax.bar(levels,count)
plt.show()
```
